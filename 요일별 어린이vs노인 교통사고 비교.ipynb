```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

dayold = pd.read_excel('dayold2.xls', index_col='', encoding='cp949')
daychild = pd.read_excel('daychild2.xls', index_col='', encoding='cp949')

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = 'Malgun Gothic'

#  
dayold2015 = dayold.loc['':'', ['', '', '','','']]
dayold2016 = dayold.loc['.1':'.1', ['', '', '','','']]
dayold2017 = dayold.loc['.2':'.2', ['', '', '','','']]
dayold2018 = dayold.loc['.3':'.3', ['', '', '','','']]
dayold2019 = dayold.loc['.4':'.4', ['', '', '','','']]

fig = plt.figure()
fig, ax = plt.subplots(2,3)
fig.set_figheight(12)
fig.set_figwidth(20)

ax[0, 0].plot(dayold2015)
ax[0, 1].plot(dayold2016)
ax[0, 2].plot(dayold2017)
ax[1, 0].plot(dayold2018)
ax[1, 1].plot(dayold2019)

ax[0, 0].legend(['','','','',''])
ax[0, 1].legend(['','','','',''])
ax[0, 2].legend(['','','','',''])
ax[1, 0].legend(['','','','',''])
ax[1, 1].legend(['','','','',''])


ax[0, 0].set_title('2015   ')
ax[0, 1].set_title('2016   ')
ax[0, 2].set_title('2017   ')
ax[1, 0].set_title('2018   ')
ax[1, 1].set_title('2019   ')

ax[0, 0].set_ylabel(' ')
ax[0, 1].set_ylabel(' ')
ax[0, 2].set_ylabel(' ')
ax[1, 0].set_ylabel(' ')
ax[1, 1].set_ylabel(' ')

plt.savefig('img\\.png')
```
