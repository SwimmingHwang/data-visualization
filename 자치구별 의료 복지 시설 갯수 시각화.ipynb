```python
nursing_2016_2019 = nursing_2016_2019.reset_index()
```

```python
nursing_dist = nursing_2016_2019.drop(nursing_2016_2019[nursing_2016_2019['자치구']=='합계'].index)
```

```python
nursing_dist = nursing_dist.drop(columns ='index')
```

```python
nursing_dist = nursing_dist.set_index('기간')
```

```python
nursing_dist = nursing_dist.drop(['기간'])
nursing_dist
```

```python
nursing_dist.dtypes
```

```python
nursing_dist = nursing_dist.astype({'합계':int})
```

```python
district = list(nursing_dist['자치구'].unique())
```

```python
f, ax = plt.subplots(nrows = 5, ncols = 5 , figsize=(20, 30))
ax = ax.flatten() 

for i in range(25):
    sns.barplot(nursing_dist[nursing_dist["자치구"]== district[i]].index,
                nursing_dist[nursing_dist["자치구"]== district[i]]['합계'],
                ax = ax[i])
    ax[i].set_ylim(0,60)
    ax[i].set(xlabel = district[i])
    
plt.savefig('img\\gu_facility.png')
```

```python
a = nursing_dist[nursing_dist.index=='2019'].groupby('자치구').sum()
```

```python
a = a.sort_values(by='합계',ascending = False)
top5 = a[:5]
top5.index
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.pie(top5,
    explode=(0.1, 0, 0, 0 ,0),
    labels=top5.index,
    shadow=True,
    # autopct='%1.2f%%',
    startangle=90,
    textprops={'fontsize': 14})

plt.title('노인 의료복지시설 Top5', fontsize=20)
plt.savefig('img\\welfare_gu_top5.png')
```

```python
low5 = a.tail(5)
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.pie(low5,
    explode=(0, 0, 0, 0 ,0.1),
    labels=low5.index,
    shadow=True,
    # autopct='%1.2f%%',
    startangle=90,
    textprops={'fontsize': 14})

plt.title('노인 의료복지시설 Low5', fontsize=20)
plt.savefig('img\\welfare_gu_low5.png')
```
